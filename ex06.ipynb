# 1. Gaussian kernel 그리기

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def GaussianKernel(v1, v2, sigma):
    res = np.exp((-(np.linalg.norm(v1-v2)**2))/(2*sigma**2))
    return res
```

```python
print np.power([2,3], 2)
print np.sum([2,3])

val = GaussianKernel(np.array([0,0]), np.array([3,5]), 1.0)
print val
```

```python
d = 100
x1 = np.linspace(0, 6, d)
x2 = np.linspace(0, 10, d)
l1 = [3, 5]

Z = np.zeros((d,d))
sigma = 1.5
for i, t1 in enumerate(x1):
    for j, t2 in enumerate(x2):
        val = GaussianKernel(np.array([t1,t2]), l1, sigma)
        Z[i,j] = val
```

```python
from mpl_toolkits import mplot3d

pX, pY = np.meshgrid(x1, x2)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(pX, pY, Z, 100, cmap='jet')
ax.set_xlabel('x1')
ax.set_xlabel('x2')
ax.set_xlabel('Z')
plt.show()

fig
```

# 2. 선형 SVM 분류
## 1) 데이터 읽기

```python
import numpy as np

from scipy.io import loadmat
data = loadmat('data1.mat')
print data

X = data['X']
y = data['y']
m = len(y)
```

```python
print X.shape, y.shape
```

## 2) 그래프 그리기

```python
pos = []
neg = []

for (i, val) in enumerate(y):
    if val == 1:
        pos.append(i)
    else:
        neg.append(i)
print pos
print neg
```

```python
import matplotlib.pyplot as plt
plt.plot(X[pos,0].reshape(-1), X[pos,1].reshape(-1), 'b+', label = 'y=1')
plt.plot(X[neg,0].reshape(-1), X[neg,1].reshape(-1), 'ro', label = 'y=0')
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend(loc='upper right')
plt.title("Data 1")
plt.show()
```

## 3) 학습

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC

svm_clf = Pipeline([
    ("scaler", StandardScaler()),
    ("linear_svc", LinearSVC(C=1, loss='hinge'))
])

svm_clf.fit(X, y)
```

```python
svm_clf.predict([[2.3, 2.7],
                [3.7, 3.9]])
```

## 4) Decision boundary

```python
x_min, x_max = X[:, 0].min() - 0.2, X[:, 0].max() + 0.2
y_min, y_max = X[:, 1].min() - 0.2, X[:, 1].max() + 0.2
h = .02
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = svm_clf.predict(np.c_[xx.ravel(), yy.ravel()])

Z = Z.reshape(xx.shape)
plt.figure(1)
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)

plt.plot(X[pos,0].reshape(-1), X[pos,1].reshape(-1), 'b+', label = 'y=1')
plt.plot(X[neg,0].reshape(-1), X[neg,1].reshape(-1), 'ro', label = 'y=0')
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend(loc='upper right')

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks(())
plt.yticks(())

plt.show()
```

# 3. 비선형 SVM 분류

```python
import numpy as np

from scipy.io import loadmat
data = loadmat('data2.mat')
print data

X = data['X']
y = data['y']
m = len(y)
```

```python
print X.shape, y.shape
```

## 2) 그래프 그리기

```python
pos = []
neg = []

for (i, val) in enumerate(y):
    if val == 1:
        pos.append(i)
    else:
        neg.append(i)
print pos
print neg
```

```python
import matplotlib.pyplot as plt
plt.plot(X[pos,0].reshape(-1), X[pos,1].reshape(-1), 'b+', label = 'y=1')
plt.plot(X[neg,0].reshape(-1), X[neg,1].reshape(-1), 'ro', label = 'y=0')
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend(loc='upper right')
plt.title("Data 2")
plt.show()
```

## 3) 학습

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

svm_clf = Pipeline([
    ("scaler", StandardScaler()),
    ("linear_svc", SVC(kernel="rbf", gamma=10, C=1))
])

svm_clf.fit(X, y)
```

## 4) Decision Boundary

```python
x_min, x_max = X[:, 0].min() - 0.2, X[:, 0].max() + 0.2
y_min, y_max = X[:, 1].min() - 0.2, X[:, 1].max() + 0.2
h = .02
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = svm_clf.predict(np.c_[xx.ravel(), yy.ravel()])

Z = Z.reshape(xx.shape)
plt.figure(1)
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)

plt.plot(X[pos,0].reshape(-1), X[pos,1].reshape(-1), 'b+', label = 'y=1')
plt.plot(X[neg,0].reshape(-1), X[neg,1].reshape(-1), 'ro', label = 'y=0')
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend(loc='upper right')

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks(())
plt.yticks(())

plt.show()
```
